## Plot championship standings data using seaborn, matplotlib and fastf1

Tutorial from [This Medium article](https://medium.com/towards-formula-1-analysis/how-to-visualize-the-formula-1-championship-in-python-using-the-ergast-api-and-seaborn-ac2f88ae7248)

```python
import pandas as pd 
import numpy as np 
import random 
import requests 

import numpy as np 
import random 
import requests 

from fastf1 import plotting 
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
"""
retrieves data from ergast api using requests
:param api_endpoint : str
:returns the data retrieved contained in a dictionary 
"""
def ergast_retrieve(api_endpoint):
    url = f'https://ergast.com/api/f1/{api_endpoint}.json'
    response = requests.get(url).json()
    
    return response['MRData']
```

#### Ergast data endpoint pattern: 
http://ergast.com/api/f1/{season}/{round}/driverStandings

To get the data from the whole season, loop through all the rounds 

```python
# number of rounds 
rounds = 21
year = 2022

# empty dataframe initialization 
general_standings = pd.DataFrame() 

# information linking driver and team 
driver_team = {}

for i in range(1, rounds + 1):
    race = ergast_retrieve(f'current/{i}/driverStandings')
    race_standings = race['StandingsTable']['StandingsLists'][0]['DriverStandings']
    
    current_round = {'round': i}
    
    # collect and store informations of each drivers 
    for i in range(len(race_standings)): 
        driver = race_standings[i]['Driver']['code']
        position = race_standings[i]['position']
        
        current_round[driver] = int(position) 
        
        driver_team[driver] = race_standings[i]['Constructors'][0]['name']
        
    general_standings = general_standings.append(current_round, ignore_index=True)
    
# set the round as the dataframe index 
general_standings = general_standings.set_index('round') 
```

```python
general_standings_melted = pd.melt(general_standings.reset_index(), ['round'])
```

```python
# plotting the data 

sns.set(rc = {'figure.figsize':(11.7, 8.27)})
fig, ax = plt.subplots()

ax.set_title (f"{year} Championship standings until round {rounds}")

for driver in pd.unique(general_standings_melted['variable']):
    sns.lineplot(
        x='round', 
        y='value', 
        data=general_standings_melted.loc[general_standings_melted['variable'] == driver], 
        color=plotting.team_color(driver_team[driver])
    )
    
# invert y-axis to have championship leader on top 
ax.invert_yaxis() 

ax.set_xticks(range(1, rounds))
ax.set_yticks(range(1, 22))

# set labels of the axes 
ax.set_xlabel("Round")
ax.set_ylabel("Championship position")

ax.grid(False) 

for line, name in zip(ax.lines, general_standings.columns.tolist()):
    y = line.get_ydata()[-1]
    x = line.get_xdata()[-1]
    
    text = ax.annotate(
        name, 
        xy=(x + 0.1 , y),
        xytext=(0, 0),
        color=line.get_color(),
        xycoords=(
            ax.get_xaxis_transform(),
            ax.get_yaxis_transform()
        ),
        textcoords="offset points"
    )
    
ax.set_facecolor("black")
```
